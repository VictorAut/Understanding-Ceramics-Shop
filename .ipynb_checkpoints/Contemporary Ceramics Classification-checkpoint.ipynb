```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTENC

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

import warnings
warnings.filterwarnings('ignore')
```

```python
pots = pd.read_csv('Pots_clean_labelled.csv')
pots.drop(['price','productcode', 'name', 'maker', 'description'], axis=1, inplace=True)
pots.head(3)
```

```python
pd.Series(pots.category).value_counts()
```

We'll need to do some resampling of our data dataset as the 'third' category is currently 5 times larger than the 'first' category.

<br>

Let's hark back to the regression problem and note that the two features that featured prominently in predicting prices was 'height' and 'length' ('length' was engineered from 'width' and 'depth', so 'width' only will be showcased here)

```python
pots.drop('category', axis=1).columns
```

```python
X = pots[['height', 'width', 'depth', 'materials', 'technique', 'firing','decoration']]
y = pots['category']
```

#### Random Oversampling

```python
ros = RandomOverSampler(sampling_strategy = 'not majority', random_state = 123)

# fit the resampler to our dataset and store it into new variables
X_sampled, y_sampled = ros.fit_resample(X, y)

pots_sampled = pd.DataFrame(data=X_sampled, columns = pots.drop('category', axis=1).columns)
pots_sampled['category'] = pd.Series(y_sampled)

print(pd.Series(pots_sampled.category).value_counts())

fig, ax = plt.subplots(1,2, figsize=(18, 8))
sns.scatterplot(x='width', y='height', data = pots, hue='category', 
                alpha = 0.5, s = 20, ax=ax[0])
sns.scatterplot(x='width', y='height', data = pots_sampled, hue='category', 
                alpha = 0.5, s = 20, ax=ax[1])
```

#### SMOTE

```python
X.columns
```

```python
# create a RandomUnderSampler that will finish with a 10:1 ratio of classes, and fix the random state
smote = SMOTENC(sampling_strategy = 'not majority', random_state = 123, categorical_features = [3,4,5,6])

# fit the resampler to our dataset and store it into new variables
X_sampled, y_sampled = smote.fit_resample(X, y)

pots_sampled = pd.DataFrame(data=X_sampled, columns = pots.drop('category', axis=1).columns)
pots_sampled['category'] = pd.Series(y_sampled)

print(pd.Series(pots_sampled.category).value_counts())

fig, ax = plt.subplots(1,2, figsize=(18, 8))
sns.scatterplot(x='width', y='height', data = pots, hue='category', 
                alpha = 0.5, s = 20, ax=ax[0])
sns.scatterplot(x='width', y='height', data = pots_sampled, hue='category', 
                alpha = 0.5, s = 20, ax=ax[1])
```

<br>

# Logistic Regression

Without any sampling, then with sampling

```python
X_ = pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(X_, y, test_size = 0.3, random_state = 123)

print(X_train.shape)

model = LogisticRegression()
model.fit(X_train, y_train)

print("Original accuracy: ", accuracy_score(y_test, model.predict(X_test)))
print("Original precision: ", precision_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
print("Original recall: ", recall_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
print("Original f1: ", f1_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 123)

smote = SMOTENC(sampling_strategy = 'not majority', random_state = 123, categorical_features = [3,4,5,6])
X_sampled, y_sampled = smote.fit_resample(X_train, y_train)

X_sampled = pd.get_dummies(X_sampled)
X_test = pd.get_dummies(X_test)

model = LogisticRegression()
model.fit(X_sampled, y_sampled)

print("Original accuracy: ", accuracy_score(y_test, model.predict(X_test)))
print("Original precision: ", precision_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
print("Original recall: ", recall_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
print("Original f1: ", f1_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
```

# Decision Trees

Without sampling, then with sampling

```python
X_train, X_test, y_train, y_test = train_test_split(X_, y, test_size = 0.3, random_state = 123)

print(X_train.shape)

model = DecisionTreeClassifier(criterion = 'gini', max_depth = 5)
model.fit(X_train, y_train)

print("Original accuracy: ", accuracy_score(y_test, model.predict(X_test)))
print("Original precision: ", precision_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
print("Original recall: ", recall_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
print("Original f1: ", f1_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 123)

smote = SMOTENC(sampling_strategy = 'not majority', random_state = 123, categorical_features = [3,4,5,6])
X_sampled, y_sampled = smote.fit_resample(X_train, y_train)

X_sampled = pd.get_dummies(X_sampled)
X_test = pd.get_dummies(X_test)

model = DecisionTreeClassifier(criterion = 'gini', max_depth = 5)
model.fit(X_sampled, y_sampled)

print("Original accuracy: ", accuracy_score(y_test, model.predict(X_test)))
print("Original precision: ", precision_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
print("Original recall: ", recall_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
print("Original f1: ", f1_score(y_test, model.predict(X_test), labels=['first', 'second', 'third'], average=None))
```
